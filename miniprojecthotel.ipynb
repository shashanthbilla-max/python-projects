```python
def welcome():
    print("Welcome to Meni's Hotel")
    print("------------------------")

def hotel_main_menu():
    print("\nSelect your menu type:")
    print("1. TIFFIN AND TEA")
    print("2. LUNCH")
    print("3. DINNER")
    print("4. EXIT")

def tiffin_menu():
    print("\n--- Tiffin Menu ---")
    print("1. Dosa - 80")
    print("2. Puri - 70")
    print("3. Wada - 60")
    print("4. Upma - 50")
    print("5. Idly - 40")
    print("6. Exit")

def lunch_menu():
    print("\n--- Lunch Menu ---")
    print("1. Veg Biriyani - 80")
    print("2. Chicken Biriyani - 200")
    print("3. Paneer Biriyani - 170")
    print("4. Tomato Rice - 60")
    print("5. Curd Rice - 50")
    print("6. Exit")

def dinner_menu():
    print("\n--- Dinner Menu ---")
    print("1. Chapathi - 60")
    print("2. Parota - 70")
    print("3. Veg Fried Rice - 120")
    print("4. Egg Fried Rice - 140")
    print("5. Chicken Fried Rice - 160")
    print("6. Exit")
def take_order():
    total_breakfast = 0
    total_lunch = 0
    total_dinner = 0
    orders = []
    while True:
        hotel_main_menu()
        meal_choice = input("Enter your choice (1-4): ")
        if meal_choice == '1':
            while True:
                tiffin_menu()
                choice = input("Enter your choice (1-6): ")

                if choice == '1':
                    num = int(input("How many dosas? "))
                    total_breakfast += 80 * num
                    orders.append(("Dosa", num, 80 * num))
                    print(f"Added {num} dosa(s)")
                elif choice == '2':
                    num = int(input("How many puris? "))
                    total_breakfast += 70 * num
                    orders.append(("Puri", num, 70 * num))
                    print(f"Added {num} puri(s)")
                elif choice == '3':
                    num = int(input("How many wadas? "))
                    total_breakfast += 60 * num
                    orders.append(("Wada", num, 60 * num))
                    print(f"Added {num} wada(s)")
                elif choice == '4':
                    num = int(input("How many upmas? "))
                    total_breakfast += 50 * num
                    orders.append(("Upma", num, 50 * num))
                    print(f"Added {num} upma(s)")
                elif choice == '5':
                    num = int(input("How many idlys? "))
                    total_breakfast += 40 * num
                    orders.append(("Idly", num, 40 * num))
                    print(f"Added {num} idly(s)")
                elif choice == '6':
                    print(f"\nBreakfast total: ₹{total_breakfast}")
                    break
                else:
                    print("Invalid choice!")
        elif meal_choice == '2':
            while True:
                lunch_menu()
                choice = input("Enter your choice (1-6): ")

                if choice == '1':
                    num = int(input("How many veg biriyanis? "))
                    total_lunch += 80 * num
                    orders.append(("Veg Biriyani", num, 80 * num))
                    print(f"Added {num} veg biriyani(s)")
                elif choice == '2':
                    num = int(input("How many chicken biriyanis? "))
                    total_lunch += 200 * num
                    orders.append(("Chicken Biriyani", num, 200 * num))
                    print(f"Added {num} chicken biriyani(s)")
                elif choice == '3':
                    num = int(input("How many paneer biriyanis? "))
                    total_lunch += 170 * num
                    orders.append(("Paneer Biriyani", num, 170 * num))
                    print(f"Added {num} paneer biriyani(s)")
                elif choice == '4':
                    num = int(input("How many tomato rice? "))
                    total_lunch += 60 * num
                    orders.append(("Tomato Rice", num, 60 * num))
                    print(f"Added {num} tomato rice(s)")
                elif choice == '5':
                    num = int(input("How many curd rice? "))
                    total_lunch += 50 * num
                    orders.append(("Curd Rice", num, 50 * num))
                    print(f"Added {num} curd rice(s)")
                elif choice == '6':
                    print(f"\nLunch total: ₹{total_lunch}")
                    break
                else:
                    print("Invalid choice!")
        elif meal_choice == '3':
            while True:
                dinner_menu()
                choice = input("Enter your choice (1-6): ")

                if choice == '1':
                    num = int(input("How many chapathis? "))
                    total_dinner += 60 * num
                    orders.append(("Chapathi", num, 60 * num))
                    print(f"Added {num} chapathi(s)")
                elif choice == '2':
                    num = int(input("How many parotas? "))
                    total_dinner += 70 * num
                    orders.append(("Parota", num, 70 * num))
                    print(f"Added {num} parota(s)")
                elif choice == '3':
                    num = int(input("How many veg fried rice? "))
                    total_dinner += 120 * num
                    orders.append(("Veg Fried Rice", num, 120 * num))
                    print(f"Added {num} veg fried rice(s)")
                elif choice == '4':
                    num = int(input("How many egg fried rice? "))
                    total_dinner += 140 * num
                    orders.append(("Egg Fried Rice", num, 140 * num))
                    print(f"Added {num} egg fried rice(s)")
                elif choice == '5':
                    num = int(input("How many chicken fried rice? "))
                    total_dinner += 160 * num
                    orders.append(("Chicken Fried Rice", num, 160 * num))
                    print(f"Added {num} chicken fried rice(s)")
                elif choice == '6':
                    print(f"\nDinner total: ₹{total_dinner}")
                    break
                else:
                    print("Invalid choice!")
        elif meal_choice == '4':
            grand_total = total_breakfast + total_lunch + total_dinner

            print("----------FINAL BILL----------")
            print("{:<25} {:<10} {:<10}".format("Item", "Qty", "Amount"))
            print("-" * 45)
            for item, qty, amount in orders:
                print("{:<25} {:<10} ₹{:<10}".format(item, qty, amount))
            print("-" * 45)
            print(f"Total Breakfast Bill: ₹{total_breakfast}")
            print(f"Total Lunch Bill: ₹{total_lunch}")
            print(f"Total Dinner Bill: ₹{total_dinner}")
            print("---------------------------------------------")
            print(f"GRAND TOTAL: ₹{grand_total}")
            print("---------------------------------------------")
            name = input("your name sir:")
            print("Thank you for visiting Meni's Hotel(mr/ms):",name)
            break

        else:
            print("Invalid choice. Please try again.")
welcome()
take_order()
```
